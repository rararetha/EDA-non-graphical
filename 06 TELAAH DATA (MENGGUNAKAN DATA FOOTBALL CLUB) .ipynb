**9. Mendeteksi Outlier Menggunakan Rumus Tukey's Fences IQR (Spread)**

Data dianggap sebagai `outliers` apabila :

$$ < (Q1 - 1.5 \times IQR) $$
$$ atau $$
$$ > (Q3 - 1.5 \times IQR) $$

Rumus IQR (Interquartile) :
$$ IQR = Q3-Q1 $$

```python
q1 = dfnew.quantile(0.25) # minimum
q3 = dfnew.quantile(0.75) # maximum
iqr = q3 - q1
iqr
```

```python
outlier_criteria = (dfnew < (q1-1.5*iqr)) | (dfnew > (q3+1.5*iqr))
outlier_criteria
```

```python
outlier_criteria.sum()
```

```python
# data yang outlier 

df_outlier = dfnew[outlier_criteria.any(axis=1)]
df_outlier
```

```python
# data yang tidak outlier 

df_no_outlier = dfnew[~outlier_criteria.any(axis=1)]
df_no_outlier
```

```python
# memfilter assist dan pemain yang termasuk dalam outlier

dfnew.loc[:, ['player_name', 'assists']][outlier_criteria['assists']].sort_values(by = ['assists'], ascending = False)
```
